```python
from matplotlib import pyplot as plt
import numpy as np
shape_size = 48
def shape(batch, index):
    b = np.genfromtxt('data/DATA'+str(batch) +'_sh.txt')
    start = index * shape_size
    x = b[start:start + shape_size, 0]
    y = b[start:start + shape_size, 1]
    plt.axis('equal')
    plt.ylim(-0.5, 0.5)
    plt.xlim(-0.5, 0.5)
    plt.plot(x, y)
    plt.fill(x, y)
    plt.show()
```

```python
shape(62, 202)
```
